# 数据处理
先对三个数据集导入并进行数据处理。

```python
import numpy as np
import pandas as pd
ak=pd.read_csv(r"C:\Users\林之韵\Desktop\py\bilibili_明日方舟_热门TOP1000.xls")
```

```python
fgo=pd.read_csv(r"C:\Users\林之韵\Desktop\py\bilibili_FGO_热门TOP1000.xls")
```

```python
ys=pd.read_csv(r"C:\Users\林之韵\Desktop\py\bilibili_原神_热门TOP1000.xls")
```

```python
for i in range(0,len(ak)):
    b=ak['播放量'].loc[i].replace("\n        ","").replace("万\n      ","")
    ak['播放量'].loc[i]=b
    ak['播放量'].loc[i]=float(ak['播放量'].loc[i])
```

```python
for i in range(0,len(ys)):
    b=ys['播放量'].loc[i].replace("\n        ","").replace("万\n      ","")
    ys['播放量'].loc[i]=b
    ys['播放量'].loc[i]=float(ys['播放量'].loc[i])
```

```python
for i in range(0,len(fgo)):
    b=fgo['播放量'].loc[i].replace("\n        ","").replace("万\n      ","")
    fgo['播放量'].loc[i]=b
    fgo['播放量'].loc[i]=float(fgo['播放量'].loc[i])
```

```python
ak.时间=pd.to_datetime(ak.时间)
ys.时间=pd.to_datetime(ys.时间)
fgo.上传时间=pd.to_datetime(fgo.上传时间)
```

```python
e1=ak.视频时长.str.split(":")
e2=ys.视频时长.str.split(":")
```

```python
for i in range(0,len(e1)):
    if len(e1[i])==2:
        ak.视频时长[i]=format(float(e1[i][-1])*0.0167+float(e1[i][-2]),".2f")
    if len(e1[i])==3:
        ak.视频时长[i]=format(float(e1[i][-1])*0.0167+float(e1[i][-2])+float(e1[i][-3])*3,".2f")
```

```python
for i in range(0,len(e2)):
    if len(e2[i])==2:
        ys.视频时长[i]=format(float(e2[i][-1])*0.0167+float(e2[i][-2]),".2f")
    if len(e2[i])==3:
        ys.视频时长[i]=format(float(e2[i][-1])*0.0167+float(e2[i][-2])+float(e2[i][-3])*3,".2f")
```
