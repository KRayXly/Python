## 基本文件操作

* open
* readlines()

```python
# 打开某个文件，注意目录字符串前面加"r",最后的一个字符r代表打开的方式，r代表已只读的方式打开
op=open(r"D:\try\t_alibaba_data31.txt","r") 

# readlines()是文件操作的一个常用方法，是在已经打开文件的基础上，把文件按行读入，注意这样读入后的结果是以行为元素的列表。
a=op.readlines()#同时会关闭文件

# 可以通过循环输出部分数据,因为数据有几万行，最好不要全部输出
for i in a[0:5]:
    print(i.split("\t"))
```

```python
#注意看下面第一行的原始形状，首先是用“\t”进行分割的，这其实就是一个空格符（tab键），但是比简单的空格长一点
# 最后是“\n”符,这是换行符，数据是分行的，说明在这个位置后下面的数据会另取一行
a[0]
```

```python
#print 则是显示这些符号的效果
print(a[0])
```

```python
#通过split可以把每列的数据分割开来
#注意一般数据读入最初都是默认为字符串，即便它看起来是整数
print(a[0].split("\t"))
```

```python
#对“12154500”用户，以其所有浏览过的品牌为键，浏览次数为值构建字典
#然后以其所有购买过的品牌为键，购买次数为值构建字典
```

```python
c={}
for i in b:
    if i[2]=="0":
        c[i[1]]=c.get(i[1],0)+1
print(c)
```

```python
d={}
for i in b:
    if i[2]=="1":
        d[i[1]]=d.get(i[1],0)+1
print(d)
```

```python
#此用户浏览过多少品牌，购买过多少品牌？
```

```python
print(len(c),len(d))
```

```python
#对此用户，平均每个品牌浏览次数？平均每个品牌的购买次数？
```

```python
print(sum(c.values())/len(c),sum(d.values())/len(d))
```

```python
#此用户有哪些品牌浏览但是没购买过？（用set 求补集） 
```

```python
print(set(c.keys())-set(d.keys()))
```

```python
#求此用户没有购买过的品牌的平均的浏览次数？
```

```python
e=set(c.keys())-set(d.keys())
f=0
for i in e:
    f+=int(c[i])
print(f/len(e))
```

```python
#此用户对于购买过的商品的，输出其商品id，浏览次数，购买次数；
#并且最后求这些商品的平均浏览次数？
```

```python
f=0
for i in d:
    f+=int(c[i])
    print(i,c[i],d[i])
print(f/len(d))
    
```
