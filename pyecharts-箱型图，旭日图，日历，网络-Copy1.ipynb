* 构建一个简单的两层旭日图

* 利用上面的结果做出前四个店铺的商品数量旭日图

```python
# 求出每个店铺下不同鞋面材质，不同款式的商品数量 p2
p2=a.groupby(["nick","info.鞋面材质","info.款式"]).size()
```

```python
p2
```

```python
store=["意尔康皮鞋旗舰店","米兰多格商场","意尔康男鞋旗舰店","顺时针方向感"]
color=["#da0d68","#975e6d","#e0719c","#f99e1c","#ef5a78","#f7f1bd","#da1d23","#dd4c51","#3e0317","#e62969","#6569b0"]
```

```python
p2.index
```

```python
import random
random.choice(color)
```

```python
#这个代码的前提是index本身是要已经排序好了
data=[]
for i in store:
    z1={"name":i,"itemStyle": {"color": random.choice(color)},"children":[]}
    i0=[0,0]
    z2={}
    for j in p2[i].index:
        if j[0]!=i0[0]:
            i0[0]=j[0]
            if len(z2)>0:z1["children"].append(z2)
            z2={"name":j[0],"itemStyle": {"color": random.choice(color)},"children":[]}
            z3={"name":j[1],"itemStyle": {"color": random.choice(color)},"value":int(p2[i][j[0]][j[1]])}
            z2["children"].append(z3)
        else:
            z3={"name":j[1],"itemStyle": {"color": random.choice(color)},"value":int(p2[i][j[0]][j[1]])}
            z2["children"].append(z3)
    z1["children"].append(z2)
    data.append(z1)
```

```python
from pyecharts.charts import Sunburst
from pyecharts import options as opts
c = (
    Sunburst(init_opts=opts.InitOpts(width="1000px", height="600px"))
    .add(
        "",
        data_pair=data,
        highlight_policy="ancestor",
        radius=[0, "95%"],
        sort_="null",
        levels=[
            {},
            {
                "r0": "15%",
                "r": "35%",
                "itemStyle": {"borderWidth": 2},
                "label": {"rotate": "tangential"},
            },
            {"r0": "35%", "r": "70%", "label": {"align": "right"}},
            {
                "r0": "70%",
                "r": "72%",
                "label": {"position": "outside", "padding": 3, "silent": False},
                "itemStyle": {"borderWidth": 3},
            },
        ],
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Sunburst-官方示例"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}"))
)
c.render_notebook()
#参考此例子https://gallery.pyecharts.org/#/Sunburst/drink_flavors
```

## 日历图



```python
#https://gallery.pyecharts.org/#/Calendar/calendar_base
import datetime
import random
from pyecharts import options as opts
from pyecharts.charts import Calendar


begin = datetime.date(2017, 1, 1)
end = datetime.date(2017, 12, 31)

data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
    for i in range((end - begin).days + 1)
]

C1=Calendar()
C1.add("", data, calendar_opts=opts.CalendarOpts(range_="2017"))
C1.set_global_opts(
        title_opts=opts.TitleOpts(title="Calendar-2017年微信步数情况"),
        visualmap_opts=opts.VisualMapOpts(
            max_=20000,
            min_=500,
            orient="horizontal",
            is_piecewise=True,
            pos_top="230px",
            pos_left="100px",
        ),
    )
C1.render_notebook()
```

```python
from pyecharts.charts import Graph
```

```python
#https://gallery.pyecharts.org/#/Graph/graph_base

nodes = [
    {"name": "结点1", "symbolSize": 10},
    {"name": "结点2", "symbolSize": 20},
    {"name": "结点3", "symbolSize": 30},
    {"name": "结点4", "symbolSize": 40},
    {"name": "结点5", "symbolSize": 50},
    {"name": "结点6", "symbolSize": 40},
    {"name": "结点7", "symbolSize": 30},
    {"name": "结点8", "symbolSize": 20},
]
links = []
for i in nodes:
    for j in nodes:
        links.append({"source": i.get("name"), "target": j.get("name")})
G1=Graph()
G1.add("", nodes, links, repulsion=8000)
G1.set_global_opts(title_opts=opts.TitleOpts(title="Graph-基本示例"))
G1.render_notebook()
```

## classwork4

* 构建数据，用graph输出一个三角形

* 提取"2011-07-04"到"2011-07-10"的购买记录，品牌与用户为点，用户购买品牌为边，构建网络图

```python
links=[]
user=set([])
brand=set([])
for i in r1[(r1.date>"2011-07-04")&(r1.behavr==1)&(r1.date<"2011-07-10")].groupby(["brand","user"]).size().iteritems():
    links.append({"source":str(i[0][0]),"target":str(i[0][1])})
    user.add(str(i[0][0]))
    brand.add(str(i[0][1]))
```

```python
nodes=[]
for i in user:
    nodes.append({"name":i,"category":0})
for i in brand:
    nodes.append({"name":i,"category":1})
#nodes
```

```python
G2 = Graph()
G2.add("", nodes, links,[{"name":"user"},{"name":"items"}])
G2.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
G2.render_notebook()
```
