```python
g1=ak[ak.作者=='哔哩哔哩游戏中心'].groupby("播放量").size()
g2=ys[ys.作者=='哔哩哔哩直播'].groupby("播放量").size()
g3=fgo[fgo.作者=='哔哩哔哩游戏中心'].groupby("播放量").size()
g1
```

```python
data1=[]
data2=[]
data3=[]
s=0
t=1
for i in g1.items():
    s+=i[0]
    t+=1
data1.append(s)
data2.append(t)
data3.append("明日方舟")
s=0
t=1
for i in g3.items():
    s+=i[0]
    t+=1
data1.append(s)
data2.append(t)
data3.append("FGO")
s=0
t=1
for i in g2.items():
    s+=i[0]
    t+=1
data1.append(s)
data2.append(t)
data3.append("原神")
```

```python
z1=[list(z) for z in zip(data3,data1)]
z2=[list(z) for z in zip(data3,data2)]
```

```python
import pyecharts.options as opts
from pyecharts.charts import Pie
(
    Pie()
    .add(
        series_name="B站官方推荐视频数",
        data_pair=z1,
        radius=[0, "40%"],
        label_opts=opts.LabelOpts(position="inner",),
    )
    .add(
        series_name="官方推荐播放总量",
        radius=["50%", "75%"],
        data_pair=z2,
        rosetype="radius",
        label_opts=opts.LabelOpts(position="outside",formatter="{a} : {d}%"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="B站官方推荐视频数据对比",pos_left="left",pos_top="30%"),
        legend_opts=opts.LegendOpts(pos_left="left",pos_top="middle",orient="vertical"),
    )
    .set_colors([
                "#abd9e9",
                "#fee090",
                "#fdae61",
            ]
    )
    .render_notebook()
)
```
